```python
import sys
import os

path = "/content/drive/MyDrive/NLP_Project_New"
sys.path.append(os.path.abspath(path))
```

```python
import pandas as pd
from faker import Faker
```

```python
test_df = pd.read_csv(f'{path}/data/train_test_val/test.tsv', sep="\t", skip_blank_lines=False, names = ["TOKEN", "NE"])
```

```python
test_df = test_df.head(54303)
```

```python
adv_tests = []
```

```python
# perturb_1: Replace all Names with one random string
perturb_1 = test_df.copy()
perturb_1["TOKEN"] = perturb_1.apply(lambda row: "Dodo" if row["NE"] in ("B-PER", "I-PER") else row["TOKEN"], axis=1)
adv_tests.append(perturb_1)
```

```python
fake = Faker('de_DE')
```

```python
# perturb_2: replace all person names with German names
perturb_2 = test_df.copy()
perturb_2["TOKEN"] = perturb_2.apply(lambda row: "-".join(fake.first_name().split()).replace(".", "").replace("’", "") if row["NE"] == "B-PER" else row["TOKEN"], axis=1)
perturb_2["TOKEN"] = perturb_2.apply(lambda row: "-".join(fake.last_name().split()).replace(".", "").replace("’", "") if row["NE"] == "I-PER" else row["TOKEN"], axis=1)
adv_tests.append(perturb_2)
```

```python
# perturb_3: replace all place names with German place names
perturb_3 = test_df.copy()
perturb_3["TOKEN"] = perturb_3.apply(lambda row: "-".join(fake.city_name().split()).replace(".", "").replace("’", "") if row["NE"] in ("B-LOC", "I-LOC") else row["TOKEN"], axis=1)
adv_tests.append(perturb_3)
```

```python
# perturb_4: replace all person and place names
# with German person and place names
perturb_4 = perturb_2.copy()
perturb_4["TOKEN"] = perturb_4.apply(lambda row: "-".join(fake.city_name().split()).replace(".", "").replace("’", "") if row["NE"] in ("B-LOC", "I-LOC") else row["TOKEN"], axis=1)
adv_tests.append(perturb_4)
```

```python
fake = Faker('en_IN')
```

```python
# perturb_5: replace all person names with Indian person names
perturb_5 = test_df.copy()
perturb_5["TOKEN"] = perturb_5.apply(lambda row: "-".join(fake.first_name().split()).replace(".", "").replace("’", "") if row["NE"] == "B-PER" else row["TOKEN"], axis=1)
perturb_5["TOKEN"] = perturb_5.apply(lambda row: "-".join(fake.last_name().split()).replace(".", "").replace("’", "") if row["NE"] == "I-PER" else row["TOKEN"], axis=1)
adv_tests.append(perturb_5)
```

```python
# perturb_6: replace all place names with Indian place names
perturb_6 = test_df.copy()
perturb_6["TOKEN"] = perturb_6.apply(lambda row: "-".join(fake.city_name().split()).replace(".", "").replace("’", "") if row["NE"] in ("B-LOC", "I-LOC") else row["TOKEN"], axis=1)
adv_tests.append(perturb_6)
```

```python
# perturb_7: replace all person and place names
# with Indian person and place names
perturb_7 = perturb_5.copy()
perturb_7["TOKEN"] = perturb_7.apply(lambda row: "-".join(fake.city_name().split()).replace(".", "").replace("’", "") if row["NE"] in ("B-LOC", "I-LOC") else row["TOKEN"], axis=1)
adv_tests.append(perturb_7)
```
